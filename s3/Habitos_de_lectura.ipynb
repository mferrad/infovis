````python
import altair as alt
alt.renderers.enable('default')
````

````python
import pandas as pd

allBooksdf = pd.read_csv("https://raw.githubusercontent.com/mferrad/infovis/main/s3/goodreads_library_export.csv")
````

````python
readBooks=allBooksdf.where(allBooksdf["Read_Count"]>0)
readBooks.head()
````

````python
source = readBooks["Number_of_Pages"].dropna().reset_index()

base = alt.Chart(source)

bar = base.mark_bar().encode(
    x=alt.X("Number_of_Pages:Q", bin=alt.Bin(extent=[-0, 1400], step=100),axis=alt.Axis(title="Numero de paginas", labelAngle=0)),
    y=alt.Y('count()',axis=alt.Axis(title="Cantidad", labelAngle=0))
)

rule = base.mark_rule(color='red').encode(
    x='mean(Number_of_Pages):Q',
    size=alt.value(5)
)

bar + rule
````

````python
source = readBooks["Original_Publication_Year"].dropna().reset_index()

alt.Chart(source).mark_bar().encode(
    x=alt.X("Original_Publication_Year:O",axis=alt.Axis(title="Fecha de publicacion")),
    y="count()",
    # The highlight will be set on the result of a conditional statement
    color=alt.condition(
        alt.datum.Original_Publication_Year== 2006,  # If the year is 1810 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(width=1500)
````

````python
source = readBooks["Original_Publication_Year"].dropna().reset_index()

base = alt.Chart(source)

base=base.mark_bar().encode(
    x=alt.X("Original_Publication_Year:O", bin=alt.Bin(extent=[-500, 2022], step=20),axis=alt.Axis(title="Fecha de publicacion", labelAngle=0)),
    y=alt.Y('count()',axis=alt.Axis(title="Cantidad", labelAngle=0))
).properties(width=1000)

base
````

````python
readBooksWithDate=readBooks[readBooks['Date_Read'].notna()]
readBooksWithDate=readBooksWithDate[['Date_Read','Read_Count']]
readBooksWithDate=readBooksWithDate.groupby(['Date_Read']).sum().reset_index()  #elimino las repeticiones
readBooksWithDate['Date_Read']=readBooksWithDate['Date_Read'].astype('datetime64[ns]')
readBooksWithDate.sort_values(by='Date_Read',inplace=True)
readBooksWithDate['count'] = readBooksWithDate.Read_Count.cumsum()
readBooksWithDate.rename(columns = {'Date_Read':'Date'}, inplace = True)
readBooksWithDate
````

````python
source = readBooksWithDate

alt.Chart(source).mark_line().transform_window(
    # Sort the data chronologically
    sort=[{'field': 'Date'}],
    # Include all previous records before the current record and none after
    # (This is the default value so you could skip it and it would still work.)
    frame=[None, 0],
    # What to add up as you go
    cumulative_read='sum(Read_Count)'
).encode(
    x='Date:T',
    # Plot the calculated field created by the transformation
    y='cumulative_read:Q'
).properties(width=1000)
````

````python
notReadBooks=allBooksdf.where(allBooksdf["Read_Count"]==0)
AddedBooksWithDate=notReadBooks[notReadBooks['Date_Added'].notna()]
AddedBooksWithDate=AddedBooksWithDate[['Date_Added']]
AddedBooksWithDate['Added_flag']=1
AddedBooksWithDate=AddedBooksWithDate.groupby(['Date_Added']).sum().reset_index()  #elimino las repeticiones
AddedBooksWithDate['Date_Added']=AddedBooksWithDate['Date_Added'].astype('datetime64[ns]')
AddedBooksWithDate.sort_values(by='Date_Added',inplace=True)
AddedBooksWithDate['count'] = AddedBooksWithDate.Added_flag.cumsum()
AddedBooksWithDate.rename(columns = {'Date_Added':'Date'}, inplace = True)
AddedBooksWithDate
````

````python
source = AddedBooksWithDate

alt.Chart(source).mark_line().transform_window(
    # Sort the data chronologically
    sort=[{'field': 'Date'}],
    # Include all previous records before the current record and none after
    # (This is the default value so you could skip it and it would still work.)
    frame=[None, 0],
    # What to add up as you go
    cumulative_read='sum(Added_flag)'
).encode(
    x='Date:T',
    # Plot the calculated field created by the transformation
    y='cumulative_read:Q'
).properties(width=1000)
````

````python
aux1=AddedBooksWithDate[['Date','count']]
aux1['Label']='Net amount added to the TBR'
aux2=readBooksWithDate[['Date','count']]
aux2['Label']='Number of books read'
junta = [aux1, aux2]
result = pd.concat(junta)
result
````

````python
source = result

alt.Chart(source).mark_area(opacity=0.3).encode(
    x="Date:T",
    y=alt.Y("count:Q", stack=None),
    color="Label:N"
)
````

````python
n=20

autoresLeidos=readBooks[readBooks['Author'].notna()]
autoresLeidos=autoresLeidos[['Author','Read_Count']]
autoresLeidos=autoresLeidos.groupby(['Author']).sum().reset_index() 
autoresLeidos.sort_values(by='Read_Count',inplace=True,ascending=False)
autoresLeidos=autoresLeidos.head(n)
autoresLeidos.to_csv('Autores.csv', index=False)
````

````python
raiting=readBooks[readBooks['My_Rating'].notna()]
raiting=readBooks[readBooks['My_Rating']>0]
raiting=raiting[['My_Rating','Average_Rating','Read_Count']]
raiting['Average_Rating']=raiting['Average_Rating'].round()
raiting=raiting.groupby(['Average_Rating','My_Rating']).sum().reset_index() 
raiting[['Average_Rating','My_Rating']]=raiting[['Average_Rating','My_Rating']].astype('int').astype('string')  
raiting.replace(to_replace="3",value="3 stars", inplace=True)
raiting.replace("1", "1 stars", inplace=True)
raiting.replace("2", "2 stars", inplace=True)
raiting.replace("4", "4 stars", inplace=True)
raiting.replace("5", "5 stars", inplace=True)
raiting
raiting.to_csv('rating.csv', index=False)
````

````python
numberOfPagesRead=readBooks['Number_of_Pages'].sum()
timeSpentReading=numberOfPagesRead*1.5
````

````python
SuperficieCanchaDeBasquet= 28 * 15 *100 *100
GameOfthronesDuration=70*60+14
SuperficiePagina=15*23
PaginasBiblia=1200

````

````python
#cantidad de Monumentales cubiertas
CantCanchasCubiertas=numberOfPagesRead*SuperficiePagina*0.5/SuperficieCanchaDeBasquet
CantCanchasCubiertas=CantCanchasCubiertas.round()
CantCanchasCubiertas
````

````python
#equivalente en biblias leidas

cantidadDeBibliasLeidas=numberOfPagesRead/PaginasBiblia
cantidadDeBibliasLeidas=cantidadDeBibliasLeidas.round()
cantidadDeBibliasLeidas
````

````python
#Cantidad de veces que vi GOT

cantidadDeVecesQueViGot=timeSpentReading/GameOfthronesDuration
cantidadDeVecesQueViGot=cantidadDeVecesQueViGot.round()
cantidadDeVecesQueViGot
````

````python
MyList=list()

for i in range(0,int(CantCanchasCubiertas)):
  MyList.append({'Label': 'Canchas de basquet cubiertas', 'icono': 'pelota'}) 

for i in range(0,int(cantidadDeVecesQueViGot)):
  MyList.append({'Label': 'Vistas de GOT', 'icono': 'lobo'})

for i in range(0,int(cantidadDeBibliasLeidas)):
  MyList.append({'Label': 'Biblias leidas', 'icono': 'iglesia'})
````

````python
source = pd.DataFrame(MyList)


alt.Chart(source).mark_text(size=45, baseline='middle').encode(
    alt.X('x:O', axis=None),
    alt.Y('icono:O', axis=None),
    alt.Row('Label:N', header=alt.Header(title='')),
    alt.Text('emoji:N')
).transform_calculate(
    emoji="{'pelota': '🏀', 'iglesia': '⛪', 'lobo': '🐺'}[datum.icono]"
).transform_window(
    x='rank()',
    groupby=['Label', 'icono']
).properties(width=550, height=140)
````
