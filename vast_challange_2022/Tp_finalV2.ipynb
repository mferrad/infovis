```python
aux=Participant0[['date','currentMode','minutos']].groupby(['date','currentMode']).sum().reset_index()
aux
```

```python
source = aux

alt.Chart(source).mark_area(opacity=0.3).encode(
    x="date:T",
    y=alt.Y("minutos:Q", stack=None),
    color="currentMode:N"
).properties(height=400, width=800)
```

```python
#Este grafico dibuja agrupando por año y mes

source = aux

selection = alt.selection_multi(fields=['currentMode'], bind='legend')

alt.Chart(source).mark_area().encode(
    alt.X('yearmonth(date):T', axis=alt.Axis(domain=False, format='%Y', tickSize=0)),
    alt.Y('sum(minutos):Q', stack='center', axis=None),
    alt.Color('currentMode:N', scale=alt.Scale(scheme='category20b')),
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).add_selection(
    selection
).properties(height=400, width=800)
```

```python
source = aux

alt.Chart(source).transform_filter(
    alt.datum.symbol != 'GOOG'
).mark_area().encode(
    x='date:T',
    y='minutos:Q',
    color='currentMode:N',
    row=alt.Row('currentMode:N')
).properties(height=100, width=700)
```

```python
#No entiendo el starting y ending balance

participantTravel=TravelJournaldf.where(TravelJournaldf["participantId"]==participant).dropna().reset_index()
del participantTravel['index']
participantTravel['travelStartTime']=participantTravel['travelStartTime'].astype('datetime64')
participantTravel['travelEndTime']=participantTravel['travelEndTime'].astype('datetime64')
participantTravel['checkInTime']=participantTravel['checkInTime'].astype('datetime64')
participantTravel['checkOutTime']=participantTravel['checkOutTime'].astype('datetime64')
participantTravel['balanceDelta']=participantTravel['endingBalance']-participantTravel['startingBalance']
participantTravel['Travelminutes']=(participantTravel['travelEndTime']-participantTravel['travelStartTime'])/pd.Timedelta('60s')
participantTravel['Checkminutes']=(participantTravel['checkOutTime']-participantTravel['checkInTime'])/pd.Timedelta('60s')
participantTravel['date']=participantTravel['travelStartTime'].dt.date.astype('datetime64')
participantTravel
```

```python
aux=participantTravel[['date','purpose','Travelminutes']].groupby(['date','purpose']).sum().reset_index()
aux
```
